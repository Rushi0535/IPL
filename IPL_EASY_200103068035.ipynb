```python
import pandas as pd
```

```python
ipl_data = 'All_IPL_Match_Scores_raw.xlsx'
ipl_df = pd.read_excel(ipl_data)
ipl_df
```

```python
ipl_df.Year.unique()
```

```python
clip = ipl_df.groupby('Year')
year_clip = clip['Match_Number'].count()
year_clip
```

```python
Number_of_Team_batting_first_won = ipl_df[ipl_df['Winning_Team'] == 'FirstBatting']
Number_of_Team_batting_first_won
```

```python
Number_of_Team_batting_first_won_fliter = Number_of_Team_batting_first_won.groupby('Year')
Number_of_Team_batting_first_won_fliter.Match_Number.count() 
```

```python
won_team_bat = pd.concat([year_clip,Number_of_Team_batting_first_won_fliter.Match_Number.count()],axis=1)
won_team_bat
```

```python
Number_of_Team_balling_first_won_fliter = ipl_df[ipl_df['Winning_Team'] == 'Chasing'].groupby('Year')
won_team_ball = pd.concat([won_team_bat,Number_of_Team_balling_first_won_fliter.Match_Number.count()],axis =1)
won_team_ball 
```

```python
Number_of_Team_tied_fliter = ipl_df[ipl_df['Winning_Team'] =='Match Tied'].groupby('Year')
match_tied = pd.concat([won_team_ball,Number_of_Team_tied_fliter.Match_Number.count()],axis=1)
match_tied
```

```python
match_tied_drop = match_tied.fillna(0,downcast='infer')
match_tied_drop.columns = ['Total Number of Matches','Number of Match team Batting Fisrt won','Number of Matches team Chasing Won','Number of times Match tied']
match_tied_drop
```
